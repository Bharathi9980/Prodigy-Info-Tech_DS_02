```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)


print("First 5 rows of the dataset:")
print(titanic.head())

# Step 3: Data Cleaning


print("\nMissing values in the dataset:")
print(titanic.isnull().sum())


print("\nPercentage of missing values in the dataset:")
print((titanic.isnull().sum() / len(titanic)) * 100)




titanic['Age'].fillna(titanic['Age'].median(), inplace=True)


titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)


titanic.drop('Cabin', axis=1, inplace=True)


titanic['Survived'] = titanic['Survived'].astype('category')
titanic['Pclass'] = titanic['Pclass'].astype('category')
titanic['SibSp'] = titanic['SibSp'].astype('category')
titanic['Parch'] = titanic['Parch'].astype('category')


print("\nData types after cleaning:")
print(titanic.dtypes)




plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=titanic, palette='Set1')
plt.title('Survival Distribution')  # Fixed missing closing quote
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=titanic, palette='Set2')
plt.title('Passenger Class Distribution')  # Fixed missing closing quote
plt.show()


plt.figure(figsize=(10,6))
sns.histplot(titanic['Age'], kde=True, bins=30, color='skyblue')
plt.title('Age Distribution')  # Fixed missing closing quote
plt.xlabel('Age')
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=titanic, palette='Paired')
plt.title('Survival Rate by Passenger Class and Gender')
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.violinplot(x='Survived', y='Age', data=titanic, palette='Set1')
plt.title('Age Distribution by Survival')
plt.show()
```

```python

corr = titanic[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']].corr()

# Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python

g = sns.FacetGrid(titanic, col='Survived', height=6, aspect=1.2)
g.map(sns.scatterplot, 'Age', 'Fare', alpha=0.7)
g.add_legend()
plt.suptitle('Age vs. Fare by Survival', y=1.02)
plt.show()
```

```python

plt.figure(figsize=(6,4))
sns.countplot(x='Embarked', data=titanic, palette='Set2')
plt.title('Passenger Count by Embarked Location')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()
```

```python

sns.pairplot(titanic[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='Set1')
plt.suptitle('Pairplot of Numerical Variables by Survival', y=1.05)
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.boxplot(x='Pclass', y='Fare', data=titanic, palette='Set3')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.kdeplot(data=titanic, x='Age', hue='Survived', fill=True, palette='Set1')
plt.title('KDE Plot of Age Distribution by Survival')
plt.show()
```
